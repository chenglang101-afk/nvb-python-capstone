```python
# With my program I am trying to find out what statistic is most correlated with finding employment?
# Your GPA, your credits, or the # of clubs your in.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Replace 'YOUR_FILE_ID' with the actual file ID of your CSV from Google Drive.
# Make sure the file is shared with 'Anyone with the link can view'.
url = 'https://drive.google.com/uc?export=download&id=15NZhY30mK1ysQhTWatWv87oahSfOtq9y'
df = pd.read_csv(url)
print(f"Loaded: {df.shape[0]} rows, {df.shape[1]} columns")
```

```python
print(f"Shape: {df.shape}")
print()
print("Column types:")
print(df.dtypes)
print()
print("First 5 rows:")
print(df.head())
print()
print("Basic statistics:")
print(df.describe().round(2))
```

```python
gpa_employment = df.groupby('employment_status')['cumulative_gpa'].median()
print(gpa_employment)


club_employment= df.groupby('employment_status')['clubs_count'].median()
print(club_employment)

credits_employment= df.groupby('employment_status')['credits_completed'].median()
print(credits_employment)


```

```python
plt.figure(figsize=(8, 6))
gpa_employment = df.groupby('employment_status')['cumulative_gpa'].median()

plt.bar(gpa_employment.index, gpa_employment.values, color="#1f77b4")

plt.title("Median Cumulative GPA by Employment Status", fontsize=14, fontweight="bold")
plt.xlabel("Employment Status")
plt.ylabel("Median Cumulative GPA")

# Fix: Ensure the number of tick locations matches the number of labels
plt.xticks(ticks=range(len(gpa_employment.index)), labels=gpa_employment.index)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

club_employment= df.groupby('employment_status')['clubs_count'].median()

plt.bar(club_employment.index, club_employment.values, color="#1f77b4")

plt.title("Median Cumulative GPA by Average # of Clubs", fontsize=14, fontweight="bold")
plt.xlabel("Employment Status")
plt.ylabel("Median Cumulative GPA")

# Fix: Ensure the number of tick locations matches the number of labels
plt.xticks(ticks=range(len(gpa_employment.index)), labels=gpa_employment.index)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
credits_employment= df.groupby('employment_status')['credits_completed'].median()

plt.bar(credits_employment.index, credits_employment.values, color="#1f77b4")

plt.title("Median Cumulative GPA by Employment Status", fontsize=14, fontweight="bold")
plt.xlabel("Employment Status")
plt.ylabel("Median Cumulative GPA")

# Fix: Ensure the number of tick locations matches the number of labels
plt.xticks(ticks=range(len(gpa_employment.index)), labels=gpa_employment.index)

plt.tight_layout()
plt.show()
```

In summary, I found out that the most applicable correlated statistics to employment is GPA and Credits. Those with few credits were unable to find employment, likely due to lack of experience in core classes. Those with the highest GPA went for grad classes while those with the second highest GPA were employed, meanwhile those who had the lowest GPA were either searching or not unemployed. The average number of clubs a person is in is uncorrelated to employment status according to the stats.
